```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("../dataset/updated_pollution_dataset.csv")
plt.scatter(df["PM2.5"], df["PM10"])
plt.xlabel("PM2.5")
plt.ylabel("PM10")
plt.title("PM2.5 vs PM10")
plt.show()
```

```python
corr = df.drop("Air Quality", axis = 1).corr()
corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.drop("Air Quality", axis=1).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
