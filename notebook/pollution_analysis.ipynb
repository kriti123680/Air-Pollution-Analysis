```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("../dataset/updated_pollution_dataset.csv") 
df.groupby("Air Quality")["NO2"].mean().plot(kind="bar")
plt.title("Average NO2 by Air Quality")
plt.show()
```

```python
plt.scatter(df["PM2.5"], df["PM10"])
plt.xlabel("PM2.5")
plt.ylabel("PM10")
plt.title("PM2.5 vs PM10")
plt.show()
```
